#Difference Between User-Based and Item-Based Collaborative Filtering
User-Based Collaborative Filtering finds users with similar preferences and recommends items liked by those similar users.

It works by calculating similarity between users using metrics like cosine similarity or Pearson correlation.

Struggles with the cold start problem when new users have little data.

Item-Based Collaborative Filtering finds items that are similar and recommends them to users based on past interactions.

It is more stable than user-based filtering, as item preferences don’t change frequently.

# What is Collaborative Filtering and How Does It Work?
Collaborative Filtering is a recommendation technique that predicts user preferences by analyzing behavior patterns of similar users.

It works by collecting user-item interactions, computing similarity scores, and recommending items based on this data.

Can be Memory-Based (using similarity measures) or Model-Based (using ML techniques like Matrix Factorization).

Struggles with the cold start problem when there is limited data for new users/items.

Used in recommendation systems like Netflix, Amazon, and Spotify to suggest relevant content.
